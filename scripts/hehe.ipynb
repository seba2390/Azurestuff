```python
from typing import List
import numpy as np

def controlled_pauli_z_matrix(num_qubits: int, control_qubits: List[int], target_qubit: int) -> np.ndarray:
    matrix = np.identity(2**num_qubits, dtype=complex)
    
    for i in range(2**num_qubits):
        binary = format(i, '0' + str(num_qubits) + 'b')
        if all(binary[-1 - q] == '1' for q in control_qubits):
            matrix[i, i] -= 2 * (binary[-1 - target_qubit] == '1')
    return matrix
```
